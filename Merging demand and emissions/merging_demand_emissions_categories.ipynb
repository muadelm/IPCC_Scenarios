# Merging Demand and Emissions 

```python
import pandas as pd
```

```python
demand = pd.read_excel('merging_Demand.xlsx')

emissions_cat = pd.read_excel('merging_categories_emissions.xlsx')
```

```python
demand.head()
```

```python
emissions_cat.head()
```

```python
# calculate category column in the demand dataset

result = pd.merge(demand, emissions_cat, how="left", on=["Model", "Scenario"])
```

```python
result.head()
```

```python
# check the resulting shape of the datasets

print('result', ' ' , result.shape)
print('demand', ' ' , demand.shape)
print('emissions', ' ' , emissions_cat.shape)
```
